# Task
Load, merge, normalize, clean, and deduplicate JSON data from "/content/data_folder", retaining section references. Split the data into 80% training and 20% validation sets. Save the cleaned data to "lawbot_cleaned.jsonl" and provide a short preprocessing report.

## Load and merge data

### Subtask:
Load the data from the specified JSON files in `/content/data_folder` and merge them into a single dataset.


**Reasoning**:
The first step is to load the data from the JSON files, which requires importing the `json` and `os` libraries and then iterating through the files in the specified directory to read and merge the data.



```python
import json
import os

all_data = []
folder_path = '/content/data_folder'

for filename in os.listdir(folder_path):
    if filename.endswith('.json'):
        file_path = os.path.join(folder_path, filename)
        with open(file_path, 'r') as f:
            data = json.load(f)
            all_data.extend(data)

print(f"Successfully loaded and merged {len(all_data)} records.")
```

## Normalize data

### Subtask:
Transform the data into the specified instruction format: `{"instruction": <question>, "output": <answer>, "source": ["IPC" | "CrPC" | "Constitution"]}`.


**Reasoning**:
Iterate through the merged data and transform each item into the specified instruction format, storing the results in `normalized_data`.



```python
normalized_data = []
for item in all_data:
    normalized_item = {
        "instruction": item.get("question", ""),
        "output": item.get("answer", ""),
        "source": [item.get("source", "Unknown")]
    }
    normalized_data.append(normalized_item)

print(f"Successfully normalized {len(normalized_data)} records.")
```

## Clean and deduplicate data

### Subtask:
Remove duplicate entries and clean the data, ensuring section references are retained.


**Reasoning**:
Remove duplicate entries based on the "instruction" field and store the unique entries in a new list.



```python
seen_instructions = set()
cleaned_data = []

for item in normalized_data:
    if item["instruction"] not in seen_instructions:
        seen_instructions.add(item["instruction"])
        cleaned_data.append(item)

print(f"Number of records remaining after deduplication: {len(cleaned_data)}")
```

## Split data

### Subtask:
Split the cleaned data into 80% training and 20% validation sets.


**Reasoning**:
Import the necessary function for splitting the data and then split the cleaned data into training and validation sets according to the specified sizes and random state.



```python
from sklearn.model_selection import train_test_split

train_data, validation_data = train_test_split(cleaned_data, test_size=0.2, random_state=42)

print(f"Training data size: {len(train_data)}")
print(f"Validation data size: {len(validation_data)}")
```

## Save cleaned data

### Subtask:
Save the cleaned and merged data to a file named `lawbot_cleaned.jsonl`.


**Reasoning**:
Write the cleaned data to a JSONL file.



```python
import json

with open('lawbot_cleaned.jsonl', 'w') as f:
    for item in cleaned_data:
        f.write(json.dumps(item) + '\n')

print("Cleaned data saved to lawbot_cleaned.jsonl")
```
